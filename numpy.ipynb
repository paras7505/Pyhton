```python
import numpy as np
```

```python
print(np.__version__)
```

```python
arr = np.array([1,2,3,4,5])
print(arr)
print(type(arr))
```

```python
arr = np.array([1,2,3,4,'a'])
print(arr)
print(type(arr))
```

dimensions


```python
arr = np.array(33)
print(arr)
# it is 0 D array
```

```python
arr = np.array([1,2,3,4,5])
print(arr)
# it is 1 D array
```

```python
arr = np.array([[1,2,3],[4,5,6]])
print(arr)
# it is 2 D array
```

```python
arr = np.array([[[1,2,3],[4,5,6]],[[1,2,3],[4,5,6]]])
print(arr)
# it is 3 D array
```

```python
arr = np.array([1,2,3,4],ndmin=10)
print(arr)
```

array indexing


```python
arr = np.array((1,2,3,4,5))
print(arr)
print(type(arr))
print(arr[0])
print(arr[1])
print(arr[2])
print(arr[3])
print(arr[4])
```

```python
arr = np.array([1,2,3,4])
for i in arr:
  print(i)
```

changing the value by its index

```python
arr = np.array([4,5,8,9,4,0,7,9])
arr[0] = 100
print(arr)
```

inserting the value

```python
arr = np.array([4,5,8,9,4,0,7,9])
result = np.insert(arr,5,100)
print(result)
```

changing the value in dimensional arrays

```python
arr = np.array([[1,2,3],[4,5,6]])
result = np.insert(arr,1,100)
print(result)
```

```python
arr = np.array([[1,2,3],[4,5,6]])
result = np.insert(arr,0,[10,20,30],axis = 0)
print(result)
```
